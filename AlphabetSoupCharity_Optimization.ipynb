## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import pandas and read the charity_data.csv from the provided cloud URL.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()

# Identify target and features
# is_successful looks like a good target
# all other columns could be features
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
application_df
```

```python
# Determine the number of unique values in each column.
unique = application_df.nunique()
unique
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
# Looking for smaller values that occur less frequently
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# Cutoff value is 156 (T9)
# Create a list of application types to replace
# use the variable name `application_types_to_replace`
application_types_to_replace = application_type_counts[application_type_counts <= 156].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_type_counts = application_df['CLASSIFICATION'].value_counts()
classification_type_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_type_counts_above_one = classification_type_counts[classification_type_counts >1]
classification_type_counts_above_one
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = classification_type_counts[classification_type_counts<=777].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
pd.get_dummies(application_df, dtype=int)
```

```python
# Split our preprocessed data into our features and target arrays
# ChatGPT helped me to ensure that all categorical variables are converted
y = application_df['IS_SUCCESSFUL']
X = pd.get_dummies(application_df.drop(columns=['IS_SUCCESSFUL']), dtype=int)


# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print(X_train.shape)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
from tensorflow.keras.callbacks import ModelCheckpoint

nn_model = tf.keras.models.Sequential()


# Input layer
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu", input_dim=43))

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu"))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu"))

# Third hidden layer
nn_model.add(tf.keras.layers.Dense(units=7, activation="relu"))

# Fourth hidden layer
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu"))

# Fifth hidden layer
nn_model.add(tf.keras.layers.Dense(units=7, activation="relu"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()

# Create callback to save model weights every 5 epochs
checkpoint_callback = ModelCheckpoint(
    filepath="model_weights.h5",
    save_weights_only =True,
    save_freq = 'epoch',
    period=5,
    verbose=1
)
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100, validation_data=(X_test, y_test), callbacks=[checkpoint_callback] )
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model.save('AlphabetSoupCharity_Optimization.h5')
```
